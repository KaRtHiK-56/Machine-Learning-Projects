# ***K MEANS CLUSTERING***

```python
x=ds.iloc[:,[1,2,3]].values
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,12):
  k=KMeans(n_clusters=i,init='k-means++',random_state=1)
  k.fit(x)
  wcss.append(k.inertia_)
plt.plot(range(1,12),wcss)

plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(1 , 12) , wcss , 'o')
plt.plot(np.arange(1 , 12) , wcss , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=6,init='k-means++',random_state=0,max_iter=300)
k_m=kmeans.fit_predict(x)
```

```python
k_m
```

```python
plt.scatter(x[k_m==0,0],x[k_m==0,1],s=100,c='red',label='1')
plt.scatter(x[k_m==1,0],x[k_m==1,1],s=100,c='blue',label='2')
plt.scatter(x[k_m==2,0],x[k_m==2,1],s=100,c='green',label='3')
plt.scatter(x[k_m==3,0],x[k_m==3,1],s=100,c='yellow',label='4')
plt.scatter(x[k_m==4,0],x[k_m==4,1],s=100,c='magenta',label='5')
plt.scatter(x[k_m==5,0],x[k_m==5,1],s=100,c='orange',label='6')
plt.show()
```

# ***K MEANS CLUSTERING***

```python
x=ds.iloc[:,[2,3]].values
x
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,12):
  k=KMeans(n_clusters=i,init='k-means++',random_state=1)
  k.fit(x)
  wcss.append(k.inertia_)
plt.plot(range(1,12),wcss)

plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(1 , 12) , wcss , 'o')
plt.plot(np.arange(1 , 12) , wcss , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=0,max_iter=300)
k_m=kmeans.fit_predict(x)
```

```python
k_m
```

```python
plt.scatter(x[k_m==0,0],x[k_m==0,1],s=100,c='red',label='1')
plt.scatter(x[k_m==1,0],x[k_m==1,1],s=100,c='blue',label='2')
plt.scatter(x[k_m==2,0],x[k_m==2,1],s=100,c='green',label='3')
plt.scatter(x[k_m==3,0],x[k_m==3,1],s=100,c='yellow',label='4')
plt.scatter(x[k_m==4,0],x[k_m==4,1],s=100,c='orange',label='5')
plt.legend()
plt.show()
```

```python
plt.scatter(x[k_m==0,0],x[k_m==0,1],s=100,c='red',label='1')
plt.scatter(x[k_m==1,0],x[k_m==1,1],s=100,c='blue',label='2')
plt.scatter(x[k_m==2,0],x[k_m==2,1],s=100,c='green',label='3')
plt.scatter(x[k_m==3,0],x[k_m==3,1],s=100,c='yellow',label='4')
plt.scatter(x[k_m==4,0],x[k_m==4,1],s=100,c='orange',label='5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='magenta',label='Centroid')
plt.xlabel("INCOME")
plt.ylabel('SCORE')
plt.legend()
plt.show()
```
